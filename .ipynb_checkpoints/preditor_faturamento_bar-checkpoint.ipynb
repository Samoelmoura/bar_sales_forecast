####3.1.5 Meta

A feature Meta é uma feature que pode dar alguns insights para buscar correlações, ela é uma feature suplementar à Target. A política de metas do bar é mensal, porém é desmembrada em metas diárias, sendo que: ãs sextas ou sábados, dias em que a demanda é maior, a meta diária é de atingir 3000 reais em vendas, já para os outros dias a meta é de 1200 reais.

Criando feature Dia_semanal

```python
class2['Dia_semanal'] = class2.index.day_name()
```

```python
def meta(x):
  d = x[0]
  v = x[1]
  if d in 'Friday Saturday'.split():
    if v >= 3000:
      return 1
    else:
      return 0
  else:
    if v >= 1000:
      return 1
    else:
      return 0
```

```python
class2['Meta'] = class2[['Dia_semanal', 'Total']].apply(meta, axis=1)
```

```python
targets.append('Meta')
```

#### Tratando valores nulos

Análises de médias, desvios padrões e teste T de student

```python
def comparar_features(series1, series2, confiança, qtd_amostras):
  print('Series 1')
  print(f'Média = {round(series1.mean(), 2)}')
  print(f'Desvio padrão = {round(series1.std(), 2)}')
  print()
  print('Series 2')
  print(f'Média = {round(series2.mean(), 2)}')
  print(f'Desvio padrão = {round(series2.std(), 2)}')
  aleatório1 = np.random.choice(series1.dropna(), qtd_amostras)
  aleatório2 = np.random.choice(series2.dropna(), qtd_amostras)
  stat, p = ttest_ind(aleatório1, aleatório2)
  if p < confiança:
    print(f'\nCom um p-valor igual à {round(p, 2)}, as series são estatísticamente DIFERENTES com um grau de confiança de {round(100-confiança*100)}%')
  if p > confiança:
    print(f'\nCom um p-valor igual à {round(p, 2)}, as series são estatísticamente IGUAIS com um grau de confiança de {round(100-confiança*100)}%')
```

##**5.1 Férias**

```python
matrix['Dia_ano'] = matrix.index.dayofyear
```

```python
dez15 = matrix[matrix.index == '2018-12-15']['Dia_ano'].values[0]
jan31 = matrix[matrix.index == '2018-01-31']['Dia_ano'].values[0]
jul01 = matrix[matrix.index == '2018-07-01']['Dia_ano'].values[0]
jul31 = matrix[matrix.index == '2018-07-31']['Dia_ano'].values[0]
```

```python
def férias(x):
  if x <= jan31:
    return 1
  elif x > jan31 & x < jul01:
    return 0
  elif x >= jul01 & x <= jul31:
    return 1
  elif x > jul31 & x < dez15:
    return 0
  else:
    return 1
```

```python
matrix['Férias'] = matrix['Dia_ano'].apply(férias)
matrix.drop('Dia_ano', axis=1, inplace=True)
```

###**6.2.2 Linear Regression com kFold**

```python
kf = KFold(n_splits=5, shuffle=True, random_state=101)
```

```python
lista_erros_absolutos = list()
lista_erros_proporcionais = list()
for train_index, test_index in kf.split(X_pca):
  X_train, X_test = X_pca.iloc[train_index], X_pca.iloc[test_index]
  y_train, y_test = y.iloc[train_index], y.iloc[test_index]
  model = LinearRegression()
  model = model.fit(X_train, y_train)
  predicts = model.predict(X_test)
  mae = np.mean(abs(y_test - predicts))
  lista_erros_absolutos.append(mae)
  mape = np.mean(abs((y_test - predicts)/y_test))
  lista_erros_proporcionais.append(mape)
```

**MAE (Média dos erros absolutos)**

```python
round(np.mean(lista_erros_absolutos), 2)
```

**MAPE (Média proporcional dos erros)**

```python
round(np.mean(lista_erros_proporcionais), 2)
```
